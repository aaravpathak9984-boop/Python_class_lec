####question
### create a matrix of 5X5

### mean of values rows and columns wise

### replace all even num with -1

##sort arrray row and column wise

### sum column wise and row wise sep

#### no. of -ve in each col

```python
# To create random matrix of 5 X 5

ma=np.random.randint(1,90,25)
matrix=ma.reshape((5,5))

print(f"The matrix is: \n {matrix} \n")

# Finding mean row wise:
mn_row=np.mean(matrix,axis=1)
print(f"The means of matrix row wise is : {mn_row} \n")

# Finding mean col wise:
mn_col=np.mean(matrix,axis=0)
print(f"The means of matrix col wise is : {mn_col} \n")


# To replace where ever we get even just replace is with -1
matrix=np.where(matrix%2==0,-1,matrix)

print(f"The matrix with replaced evens are: \n {matrix} \n")



# To sort col wise:
matrix.sort(axis=0)

# To sort row wise:
matrix.sort(axis=1)


# To find sum of rows

sum_row=np.sum(matrix,axis=1)

print(f"The sum of row wise is : \n {sum_row} \n")

#to find sum of cols:
sum_col=np.sum(matrix,axis=0)
print(f"The sum of col wise is : \n {sum_col} \n")

# to find count of total count of -1

cnt=sum(np.sum(matrix==-1,axis=0))

# can also do like:

cnt1=np.sum(np.where(matrix==-1,1,0))
print(f" The total count of -1 : \n{cnt} : {cnt1} \n")


```

### question 2

### 5X5 matrix of stock, and demand

### remaning stocks

### -ve values with 0

```python
st=np.random.randint(1,50,25)
dm=np.random.randint(1,50,25)


#stocks=st.reshape((5,5))

# can also do like:
stocks=np.random.randint(1,50,(5,5))
print(f"The stocks matrix is: \n {stocks} \n")
demand=dm.reshape((5,5))

print(f"The stocks matrix is: \n {demand} \n")



rem_stocks=stocks-demand

print(f"The remaining stocks are: \n{rem_stocks}\n")

posi_stocks=np.copy(rem_stocks)
posi_stocks[posi_stocks<0]=0

print(f"The remaining stocks are: \n{posi_stocks}\n")


# Total shortage

total_shor_days=np.sum(rem_stocks<0,axis=0)

print(f"The total no. of products the have shortages days wise are : \n{total_shor_days}\n")


total_shor_col=np.sum(np.where(rem_stocks<0,rem_stocks,0), axis=0)
print(f"The total no. of shortage in day wise is : {-1*total_shor_col}\n")


total_shortage=np.sum(total_shor_col)
print(f"The total shortage is : {-1*total_shortage}\n")




#products with zeros stokcs are

pord_zero_stock=np.sum(np.where(posi_stocks==0,1,0))

print(f"The products with zero stocks are: \n{pord_zero_stock}\n")

##sorting column wise

posi_stocks.sort(axis=0)

print(f"The sorted col wise are: \n{posi_stocks}\n")
```
