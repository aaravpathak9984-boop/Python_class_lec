```python
prices=[]
items=[]

n=int(input("Enter no. of items you want to list: " ))

for i in range(n):
  it=input("Enter item : " )
  p=int(input("Enter price of item: "))
  items.append(it)
  prices.append(p)


for i in range(n):
  print(f'\n The {items[i]} is of price : {prices[i]}\n')

mx_in=np.argmax(prices)
min_in=np.argmin(prices)


print(f'The item with maximum price is [{items[mx_in]}] and price is {prices[mx_in]}\n')
print(f'The item with minimum price is [{items[min_in]}] and price is {prices[min_in]}\n')
print(f'The average price of items is [{np.mean(prices)}]\n')
print(f'The median price of items is [{np.median(prices)}]\n')
print(f'The total price of items is [{np.sum(prices)}]\n')
print(f'The std price of items is [{np.std(prices)}]\n')
```
