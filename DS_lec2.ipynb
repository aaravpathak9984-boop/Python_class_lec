## LIST

```python
ls=[x for x in range(1,51) if(x%2==0)]

print(ls)
```

## extend() -> to insert elements in form of list as seperate elements

```python
ls.extend([4,5,6])
# remove -> removes the value from it

ls.remove(5)

#pop -> removes the value at given passed (argument -.index)

ls.pop(2)

print(ls)

print(ls.index(24))
print(ls.count(4))


ls2=[1,4,7,1,2111,0,0,-11]

k=sum(ls2)

print(k)
```

## min-to find minimum
## max- to find maximum
## sum- to find sum of list


```python
newls=[1,-1,22,44,10]

print(min(newls),max(newls),sum(newls),sum(newls,11),sum(range(1,6)))
```

## TUPLE

### Tuple Functions

```python
tp=tuple(x**2 for x in range(1,11))

print(tp)


# count

print(tp.count(4))

#min--> to find minimum

print(min(tp))

#max--> to find maximum

print(max(tp))

#sorted()--> to return new sorted tuple

print(sorted(tp))

#index()--> returns index of given

print(tp.index(81))

```
