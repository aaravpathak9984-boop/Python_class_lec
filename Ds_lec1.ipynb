## Days Check if


````python
day = input("Enter day : ").lower()

if day == "sun":
    print("Sunday")
elif day == "mon":
    print("Monday")
elif day == "tues" or day == "tue":
    print("Tuesday")
elif day == "wed":
    print("Wednesday")
elif day == "thurs" or day == "thu":
    print("Thursday")
elif day == "fri":
    print("Friday")
elif day == "sat":
    print("Saturday")
else:
    print("Invalid input")

````
