# 🐍 Python String Methods: Complete Reference Guide

This notebook covers essential built-in string methods and functions for case manipulation, whitespace removal, text splitting, search operations, and analysis.

---

## 1. Case Transformation Methods
*Note: Python strings are immutable. These methods return a **new** string and do not modify the original variable.*

### 🔹 `lower()`
* **What it returns:** A new string with all uppercase characters converted to lowercase.
* **Syntax / Usage:** `string.lower()` (takes no arguments).

```python
text = "Python 3.12"
result = text.lower()
print(f"Original: {text} | Result: {result}")
# Output: Python 3.12 -> python 3.12
```

### 🔹 `upper()`
* **What it returns:** A new string with all lowercase characters converted to uppercase.
* **Syntax / Usage:** `string.upper()` (takes no arguments).

```python
text = "Hello World"
result = text.upper()
print(f"Original: {text} | Result: {result}")
# Output: Hello World -> HELLO WORLD
```

### 🔹 `title()`
* **What it returns:** A new string where the first character of every word is uppercase, and the rest are lowercase.
* **Syntax / Usage:** `string.title()`
* *Warning:* Symbols and apostrophes can cause unexpected capitalization (e.g., "don't" becomes "Don'T").

```python
text = "the quick brown fox"
print(text.title())
# Output: The Quick Brown Fox
```

### 🔹 `capitalize()`
* **What it returns:** A new string where only the first character of the entire string is uppercase, and all others are lowercase.
* **Syntax / Usage:** `string.capitalize()`

```python
text = "hello WORLD"
print(text.capitalize())
# Output: Hello world
```

---

## 2. Space Removal Methods

### 🔹 `strip()`
* **What it returns:** A copy of the string with leading (left) and trailing (right) whitespace characters removed.
* **Syntax / Usage:** `string.strip()`. Removes spaces, tabs (`\t`), and newlines (`\n`). Pass an optional string argument to strip specific characters instead.

```python
text = "   clean me   "
print(f"'{text.strip()}'")
# Output: 'clean me'
```

### 🔹 `lstrip()`
* **What it returns:** A copy of the string with only leading (left-side) whitespace removed.
* **Syntax / Usage:** `string.lstrip()`

```python
text = "   left   "
print(f"'{text.lstrip()}'")
# Output: 'left   '
```

### 🔹 `rstrip()`
* **What it returns:** A copy of the string with only trailing (right-side) whitespace removed.
* **Syntax / Usage:** `string.rstrip()`

```python
text = "   right   "
print(f"'{text.rstrip()}'")
# Output: '   right'
```

---

## 3. Splitting and Joining Methods

### 🔹 `split()`
* **What it returns:** A `list` of substrings.
* **Syntax / Usage:** `string.split(separator, maxsplit)`. By default, it splits by any whitespace block.

```python
text = "apple,banana,cherry"
items = text.split(",")
print(items)
# Output: ['apple', 'banana', 'cherry']
```

### 🔹 `join()`
* **What it returns:** A single concatenated string.
* **Syntax / Usage:** `separator_string.join(iterable)`. This method is called **on the delimiter**, and takes a list/tuple of strings as the argument.

```python
words = ['Python', 'is', 'fun']
sentence = " ".join(words)
print(sentence)
# Output: Python is fun
```

---

## 4. Text Modification & Search Methods

### 🔹 `replace()`
* **What it returns:** A new string with occurrences of a substring replaced.
* **Syntax / Usage:** `string.replace(old_value, new_value, count)`. The `count` parameter is optional and limits the number of replacements.

```python
text = "one fish, two fish"
print(text.replace("fish", "bird"))
# Output: one bird, two bird
```

### 🔹 `find()`
* **What it returns:** An `int` representing the lowest starting index of the substring. **Returns `-1` if the text is not found.**
* **Syntax / Usage:** `string.find(substring, start, end)`

```python
text = "Coding"
print(text.find("d"))  # Output: 2
print(text.find("z"))  # Output: -1
```

---

## 5. Inspection, Counting, & Sorting Functions

### 🔹 `startswith()`
* **What it returns:** A boolean value (`True` or `False`).
* **Syntax / Usage:** `string.startswith(substring)`

```python
text = "Python"
print(text.startswith("Py"))
# Output: True
```

### 🔹 `endswith()`
* **What it returns:** A boolean value (`True` or `False`).
* **Syntax / Usage:** `string.endswith(substring)`

```python
text = "image.png"
print(text.endswith(".png"))
# Output: True
```

### 🔹 `count()`
* **What it returns:** An `int` counting non-overlapping occurrences of a substring.
* **Syntax / Usage:** `string.count(substring)`

```python
text = "banana"
print(text.count("an"))
# Output: 2
```

### 🔹 `len()`
* **What it returns:** An `int` showing the absolute number of characters in the string.
* **Syntax / Usage:** `len(string)`. **Note:** This is a global built-in function, not a string method. Do not use a dot prefix.

```python
text = "Hi"
print(len(text))
# Output: 2
```

### 🔹 `sorted()`
* **What it returns:** A new **sorted list** of individual characters, ordered by Unicode value (numbers first, uppercase, then lowercase).
* **Syntax / Usage:** `sorted(string)`. **Note:** This is a global built-in function.

```python
text = "cba"
print(sorted(text))
# Output: ['a', 'b', 'c']
```

### valid pallindrome


````python

def valid(ch: str):
    if (ch >= 'a' and ch <= 'z') or (ch >= 'A' and ch <= 'Z') or (ch >= '0' and ch <= '9'):
        return True
    return False

s = "A man, a plan, a canal: Panama"


clean_s = ""
for i in range(len(s)):
    if valid(s[i]):
        clean_s += s[i].lower()


is_pd = True
l = len(clean_s)

for i in range(0, l):
    if clean_s[i] != clean_s[l - i - 1]:
        is_pd = False
        break

print(is_pd)
````
