#### some important funcitons


```python
# 1. Generates 6 random integers from 1 (inclusive) to 12 (exclusive)
# Result: A 1D array of shape (6,) e.g., [5, 2, 11, 7, 1, 9]
arr2 = np.random.randint(1, 12, 6)

# 2. Prints the 1D random array followed by a blank new line (\n)
print(arr2, '\n')

## if we need to convert into matrix

# 3. Reshapes the 6-element array into a 2D matrix of 3 rows and 2 columns
# Your comment is correct: (3,3) fails because it needs 9 elements, but we only have 6
brr2 = arr2.reshape((3, 2))

# 4. Reshapes the same 6-element array into a 2D matrix of 2 rows and 3 columns
brr3 = arr2.reshape((2, 3))

# 5. Prints the 3x2 matrix followed by a blank line
print(brr2, '\n')

# 6. Prints the 2x3 matrix followed by a blank line
print(brr3, '\n')

# 7. Transposes brr2 (flips its rows into columns and columns into rows)
# Since brr2 was (3, 2), at2 becomes a matrix of shape (2, 3)
at2 = brr2.transpose()

# 8. Prints the transposed 2x3 matrix followed by a blank line
print(at2, '\n')

# 9. Concatenates (joins) brr3 and at2 horizontally along the columns (axis=1)
# Both matrices are shape (2, 3). Joining them side-by-side creates a shape of (2, 6)
ac2 = np.concatenate((brr3, at2), axis=1)

# 10. Prints the final horizontally stacked 2x6 matrix
print(ac2)
```

## problem of cricket scorer

```python
sc=[]

days=np.array(['monday','tuesday','wednesday','thursday','friday','saturday','sunday'])
for i in range(7):
  x=int(input(f'Enter the score of player on {days[i]}: 1'))
  sc.append(x)

scores=np.array(sc)

print(scores)





for i in range(7):

  print(f'on {days[i]} he scored : {scores[i]}')

min_idx = np.argmin(scores)
max_idx = np.argmax(scores)



print(f'the minimum score of player is {scores.min()} on {days[min_idx]}')
print(f'the maximum score of player is {max(scores)} on {days[max_idx]}')
print(f'the average score of player is {np.mean(scores)}')
print(f'the total score of player is {np.sum(scores)}')
print(f'the median score of player is {np.median(scores)}')
print(f'the standard deviation of player is {np.std(scores)}')

```
